```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("P_ES50_d_cleaned_Nov_2020.csv", parse_dates=True, index_col=0)
```

```python
import matplotlib.pyplot as plt
```

```python
data.plot()
```

```python
r_d = np.log(data/data.shift())
r_d = r_d.dropna()
r_d.head()
```

```python
r_d.plot()
```

```python
r_d["ADS.DE"].plot()
```

```python
w_eq = [1/r_d.shape[1] for i in range(r_d.shape[1])]
```

```python
r_d["1/N"] = r_d.dot(w_eq)
```

```python
r_d.describeribe()
```

```python
InvOpp = r_d.agg(["mean", "std"]).T
```

```python
InvOpp["SR"] = InvOpp["mean"]/InvOpp["std"]
```

```python
InvOpp["mean"] = InvOpp["mean"] * 252
InvOpp["std"] = InvOpp["std"] * np.sqrt(252)
InvOpp["SR"] = InvOpp["SR"] * np.sqrt(252)
```

```python
InvOpp.describe()
```

```python
InvOpp.plot(kind="scatter", x = "std", y="mean", figsize=(15,15))

for name in r_d:
    plt.annotate(name, xy=(InvOpp.loc[name,'std']+0.002, InvOpp.loc[name,'mean']+0.002))
```
